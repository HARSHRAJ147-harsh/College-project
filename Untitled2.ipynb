```python
import pandas as pd
import matplotlib.pyplot as plt
import io

# Load data
csv_data = """Date,Country,Confirmed,Recovered,Deaths
2020-03-01,India,3,0,0
2020-04-01,India,1635,150,38
2020-05-01,India,35043,8889,1147
2020-06-01,India,198706,95527,5598
2020-07-01,India,585493,347978,17400
2020-08-01,India,1695988,1094374,36611
"""
data = pd.read_csv(io.StringIO(csv_data))

# Show data
print("COVID-19 Data:\n")
print(data)

# Total cases
total_confirmed = data['Confirmed'].sum()
total_recovered = data['Recovered'].sum()
total_deaths = data['Deaths'].sum()

print("\nTotal Confirmed Cases:", total_confirmed)
print("Total Recovered Cases:", total_recovered)
print("Total Deaths:", total_deaths)

# Plot Confirmed Cases over time
plt.figure()
plt.plot(data['Date'], data['Confirmed'])
plt.xlabel("Date")
plt.ylabel("Confirmed Cases")
plt.title("COVID-19 Confirmed Cases in India")
plt.xticks(rotation=45)
plt.show()

# Plot Deaths
plt.figure()
plt.plot(data['Date'], data['Deaths'])
plt.xlabel("Date")
plt.ylabel("Deaths")
plt.title("COVID-19 Deaths in India")
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import io

# Load IPL 2024 data
ipl_csv_data = """Match_No,Team,Runs,Result
1,CSK,176,Win
1,RCB,173,Lose
2,MI,182,Win
2,GT,168,Lose
3,KKR,210,Win
3,SRH,204,Lose
4,RR,195,Win
4,LSG,180,Lose
"""
ipl_data = pd.read_csv(io.StringIO(ipl_csv_data))

# Show data
print("IPL 2024 Match Data:\n")
print(ipl_data)

# Total matches
total_matches = ipl_data['Match_No'].nunique()
print("\nTotal Matches Played:", total_matches)

# Total runs in IPL 2024
total_runs = ipl_data['Runs'].sum()
print("Total Runs Scored:", total_runs)

# Team-wise total runs
team_runs = ipl_data.groupby('Team')['Runs'].sum()
print("\nTeam-wise Total Runs:\n")
print(team_runs)

# Bar graph of team-wise runs
plt.figure()
team_runs.plot(kind='bar')
plt.xlabel("Teams")
plt.ylabel("Total Runs")
plt.title("IPL 2024 Team-wise Runs")
plt.show()
```
